```python
# Initialization
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import math

# Set font
font = {'size': 12}
plt.rc('font', **font)
```

```python
########### Question 1 - Reproduce Adam's Paper Plot ###########

# Data type
dt = [('frac_followed', np.float64), ('frac_observed', np.float64), ('frac_observed_uncertainty', np.float64)]

# Load the data
efficiency = np.loadtxt('q1_data/Efficiency.txt', dt, skiprows=1)
purity = np.loadtxt('q1_data/Purity.txt', dt, skiprows=2)

# Plot the data for efficiency
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(13, 5.5))
ax1.plot(efficiency['frac_followed'], efficiency['frac_observed'], color='black')
ax1.set_xlim(-0.05, 1.05)
ax1.set_ylim(-0.05, 1.05)
ax1.set_xlabel("Fraction of GRBs Followed Up")
ax1.set_ylabel("Fraction of high (z > 4) GRBs observed")
ax1.set_title("Efficiency", fontsize=20)

# Plot the data for purity
ax2.plot(purity['frac_followed'], purity['frac_observed'], color='black')
ax2.set_xlim(-0.05, 1.05)
ax2.set_ylim(-0.05, 1.05)
ax2.set_xlabel("Fraction of GRBs Followed Up")
ax2.set_ylabel("Percent of observed GRBs that are high z (z > 4)")
ax2.set_title("Purity", fontsize=20)

# Grey uncertainty region and other lines for efficiency plot
ax1.fill_between(efficiency['frac_followed'], 
                 efficiency['frac_observed'] + efficiency['frac_observed_uncertainty'], 
                 efficiency['frac_observed'] - efficiency['frac_observed_uncertainty'], 
                 color='lightgrey')

xmin = float('inf')
ymax1 = 1
for x, y in zip(efficiency['frac_followed'], efficiency['frac_observed']):
    if math.fabs(x - 0.2) < xmin:
        xmin = math.fabs(x - 0.2)
        ymax1 = y
ax1.plot([0.2, 0.2], [0, ymax1], color='black', linestyle=':')
ax1.plot([0, 1], [0, 1], color='black')

# Grey uncertainty region and other lines for purity plot
ax2.fill_between(purity['frac_followed'], 
                 purity['frac_observed'] + purity['frac_observed_uncertainty'], 
                 purity['frac_observed'] - purity['frac_observed_uncertainty'], 
                 color='lightgrey')

xmin = float('inf')
ymax2 = 1
for x, y in zip(purity['frac_followed'], purity['frac_observed']):
    if math.fabs(x - 0.2) < xmin:
        xmin = math.fabs(x - 0.2)
        ymax2 = y
print ymax2
ax2.plot([0.2, 0.2], [0, ymax2], color='black', linestyle=':')
ax2.plot([0, 1], [18.0/135.0, 18.0/135.0], color='black')

# Annotation for the efficiency plot
ax1.annotate('Follow up 20% of \nbursts to capture ~55% \nof high-z events', 
             xy=(0.2, ymax1), xytext=(0.45, 0.18),
             arrowprops=dict(facecolor='black', shrink=0.05), fontsize=14)
ax1.annotate('Random \nguessing', 
             xy=(0.8, 0.8), xytext=(0.7, 0.45),
             arrowprops=dict(facecolor='black', shrink=0.05), fontsize=14)

# Annotation for the purity plot
ax2.annotate('If 20% of events are \nfollowed up, ~40% of \nthem will be high-z', 
             xy=(0.2, ymax2), xytext=(0.0, 0.85),
             arrowprops=dict(facecolor='black', shrink=0.05), fontsize=14)
ax2.annotate('Random \nguessing', 
             xy=(0.61, 18.0/135.0), xytext=(0.7, 0.5),
             arrowprops=dict(facecolor='black', shrink=0.05), fontsize=14)
```

```python
########### Question 2 - New York Temperature, Google, and Yahoo! ###########

# Data types
stockData = [('date', np.uint16), ('stock', np.float64)]
tempData = [('date', np.uint16), ('temp', np.uint8)]

# Load the data
yahoo = np.loadtxt('q2_data/yahoo_data.txt', stockData, skiprows=1)
google = np.loadtxt('q2_data/google_data.txt', stockData, skiprows=1)
nyTemp = np.loadtxt('q2_data/ny_temps.txt', tempData, skiprows=1)

# Plot the data
f, ax1 = plt.subplots(figsize=(12, 8))
ax1.plot(yahoo['date'], yahoo['stock'], color='purple', label='Yahoo! Stock Value')
ax1.plot(google['date'], google['stock'], color='blue', label='Google Stock Value')
handles1, labels1 = ax1.get_legend_handles_labels()

# Plot the temperature, sharing the same x axis
ax2 = ax1.twinx()
ax2.plot(nyTemp['date'], nyTemp['temp'], color='red', linestyle='dashed', label='NY Mon. High Temp')
plt.ylim(-150, 100)
handles2, labels2 = ax2.get_legend_handles_labels()

# Draw legends, labels and title
f.suptitle('New York Temperature, Google, and Yahoo!', fontsize=20)
ax1.set_xlabel('Date(MJD)')
ax1.set_ylabel('Value(Dollars)')
ax2.set_ylabel('Temperature(F)')
plt.legend(handles1+handles2, labels1+labels2, loc=6, frameon=False)
```
