```python
# Initialization
%matplotlib inline
import os
import matplotlib.pyplot as plt
import numpy as np
import math

# Set font
font = {'size': 13}
plt.rc('font', **font)
```

```python
########### Question 1 - Reproduce Adam's Paper Plot ###########

# Data type
dt = [('frac_followed', np.float64), ('frac_observed', np.float64), 
      ('frac_observed_uncertainty', np.float64)]

# Load the data
data_fname_eff = os.path.join('q1_data', 'Efficiency.txt')
data_fname_purity = os.path.join('q1_data', 'Purity.txt')
efficiency = np.loadtxt(data_fname_eff, dt, skiprows=1)
purity = np.loadtxt(data_fname_purity, dt, skiprows=2)
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(13, 5.5)) # divide into two subplots

# ------------------------- Efficiency plot --------------------------- #

# Plot the data, label, and title
ax1.plot(efficiency['frac_followed'], efficiency['frac_observed'], color='black')
ax1.set_xlim(-0.05, 1.05)
ax1.set_ylim(-0.05, 1.05)
ax1.set_xlabel("Fraction of GRBs Followed Up")
ax1.set_ylabel("Fraction of high (z > 4) GRBs observed")
ax1.set_title("Efficiency", fontsize=20)

# Grey uncertainty region and other lines
ax1.fill_between(efficiency['frac_followed'], 
                 efficiency['frac_observed'] + efficiency['frac_observed_uncertainty'], 
                 efficiency['frac_observed'] - efficiency['frac_observed_uncertainty'], 
                 color='lightgrey')

xmin = float('inf')
ymax1 = 1
for x, y in zip(efficiency['frac_followed'], efficiency['frac_observed']):
    if math.fabs(x - 0.2) < xmin:
        xmin = math.fabs(x - 0.2)
        ymax1 = y
ax1.plot([0.2, 0.2], [-0.05, ymax1], color='black', linestyle=':')
ax1.plot([0, 1], [0, 1], color='black')

# Annotation
ax1.annotate('Follow up 20% of \nbursts to capture ~55% \nof high-z events', 
             xy=(0.2, ymax1), xytext=(0.45, 0.18),
             arrowprops=dict(facecolor='black', shrink=0.05), fontsize=14)
ax1.annotate('Random \nguessing', 
             xy=(0.8, 0.8), xytext=(0.7, 0.45),
             arrowprops=dict(facecolor='black', shrink=0.05), fontsize=14)

# ------------------------- Purity plot --------------------------- #

# Plot the data, label, and title
ax2.plot(purity['frac_followed'], purity['frac_observed'], color='black')
ax2.set_xlim(-0.05, 1.05)
ax2.set_ylim(-0.05, 1.05)
ax2.set_xlabel("Fraction of GRBs Followed Up")
ax2.set_ylabel("Percent of observed GRBs that are high-z (z > 4)")
ax2.set_title("Purity", fontsize=20)

# Grey uncertainty region and other lines
ax2.fill_between(purity['frac_followed'], 
                 purity['frac_observed'] + purity['frac_observed_uncertainty'], 
                 purity['frac_observed'] - purity['frac_observed_uncertainty'], 
                 color='lightgrey')

xmin = float('inf')
ymax2 = 1
for x, y in zip(purity['frac_followed'], purity['frac_observed']):
    if math.fabs(x - 0.2) < xmin:
        xmin = math.fabs(x - 0.2)
        ymax2 = y
ax2.plot([0.2, 0.2], [-0.05, ymax2], color='black', linestyle=':')
ax2.plot([0, 1], [18.0/135.0, 18.0/135.0], color='black')

# Annotation
ax2.annotate('If 20% of events are \nfollowed up, ~40% of \nthem will be high-z', 
             xy=(0.2, ymax2), xytext=(0.0, 0.85),
             arrowprops=dict(facecolor='black', shrink=0.05), fontsize=14)
ax2.annotate('Random \nguessing', 
             xy=(0.61, 18.0/135.0), xytext=(0.7, 0.5),
             arrowprops=dict(facecolor='black', shrink=0.05), fontsize=14)
```

```python
########### Question 2 - New York Temperature, Google, and Yahoo! ###########
from matplotlib.ticker import MultipleLocator, FormatStrFormatter

# Data types
stockData = [('date', np.uint16), ('stock', np.float64)]
tempData = [('date', np.uint16), ('temp', np.uint8)]

# Load the data
data_fname_yahoo = os.path.join('q2_data', 'yahoo_data.txt')
data_fname_google = os.path.join('q2_data', 'google_data.txt')
data_fname_ny_temps = os.path.join('q2_data', 'ny_temps.txt')
yahoo = np.loadtxt(data_fname_yahoo, stockData, skiprows=1)
google = np.loadtxt(data_fname_google, stockData, skiprows=1)
nyTemp = np.loadtxt(data_fname_ny_temps, tempData, skiprows=1)

# Ticks, ticks, ticks,
majorLocatorLeft    = MultipleLocator(100)
minorLocatorLeft    = MultipleLocator(20)
majorLocatorRight   = MultipleLocator(50)
minorLocatorRight   = MultipleLocator(10)
majorLocatorBottom  = MultipleLocator(1000)
minorLocatorBottom  = MultipleLocator(200)

# Plot the data on the left y axis
f, ax1 = plt.subplots(figsize=(12, 8))
ax1.plot(yahoo['date'], yahoo['stock'], color='purple', label='Yahoo! Stock Value')
ax1.plot(google['date'], google['stock'], color='blue', label='Google Stock Value')
plt.ylim(-20, 770)

# Handle ticks and labels for the left y axis and the x axis
ax1.tick_params(direction='in')
ax1.xaxis.set_major_locator(majorLocatorBottom)
ax1.xaxis.set_minor_locator(minorLocatorBottom) 
ax1.yaxis.set_major_locator(majorLocatorLeft)
ax1.yaxis.set_minor_locator(minorLocatorLeft) 
handles1, labels1 = ax1.get_legend_handles_labels()

# Plot the temperature, sharing the same x axis
ax2 = ax1.twinx()
ax2.plot(nyTemp['date'], nyTemp['temp'], color='red', linestyle='dashed', label='NY Mon. High Temp')
plt.xlim(48800, 55600)
plt.ylim(-150, 100)

# Ticks and labels for the right y axis
ax2.yaxis.set_major_locator(majorLocatorRight)
ax2.yaxis.set_minor_locator(minorLocatorRight) 
handles2, labels2 = ax2.get_legend_handles_labels()

# Draw legends, labels and title
f.suptitle('New York Temperature, Google, and Yahoo!', fontsize=20)
ax1.set_xlabel('Date (MJD)')
ax1.set_ylabel('Value (Dollars)')
ax2.set_ylabel('Temperature(F)')
plt.legend(handles1+handles2, labels1+labels2, loc=6, frameon=False)
```

```python
########## Draw and Save Rectangular Regions ##########
# Reference: http://stackoverflow.com/q/12052379/2465041 and Piazza

from matplotlib.patches import Rectangle

class Annotate(object):
    def __init__(self):
        self.ax = plt.gca()
        self.rect = Rectangle((0,0), 1, 1, facecolor='yellow', edgecolor='black')
        self.x0 = None
        self.y0 = None
        self.x1 = None
        self.y1 = None
        self.ax.add_patch(self.rect)
        self.ax.figure.canvas.mpl_connect('button_press_event', self.on_press)
        self.ax.figure.canvas.mpl_connect('button_release_event', self.on_release)
        self.ax.figure.canvas.mpl_connect('motion_notify_event', self.on_motion)
    
    def on_press(self, event):
        print 'press'
        self.x0 = event.xdata
        self.y0 = event.ydata    
        self.x1 = event.xdata
        self.y1 = event.ydata
        self.rect.set_width(self.x1 - self.x0)
        self.rect.set_height(self.y1 - self.y0)
        self.rect.set_xy((self.x0, self.y0))
        self.rect.set_linestyle('dashed')
        self.ax.figure.canvas.draw()
    
    def on_motion(self,event):
        if self.on_press is True:
            return
        self.x1 = event.xdata
        self.y1 = event.ydata
        self.rect.set_width(self.x1 - self.x0)
        self.rect.set_height(self.y1 - self.y0)
        self.rect.set_xy((self.x0, self.y0))
        self.rect.set_linestyle('dashed')
        self.ax.figure.canvas.draw()
  
    def on_release(self, event):
        print 'release'
        self.x1 = event.xdata
        self.y1 = event.ydata
        self.rect.set_width(self.x1 - self.x0)
        self.rect.set_height(self.y1 - self.y0)
        self.rect.set_xy((self.x0, self.y0))
        self.rect.set_linestyle('solid')
        self.ax.figure.canvas.draw()
        print self.x0,self.x1,self.y0,self.y1
        return [self.x0,self.x1,self.y0,self.y1]
```
